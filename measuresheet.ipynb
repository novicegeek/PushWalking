```python
import numpy as np
import pandas as pd
import os
root_dir = r"D:\Overseas\German Sport University Cologne\20.Course Materials\TSM11-Project&Applied Research Methods\22.Measurement Data"
os.chdir(root_dir)
```

```python
# Set parameters
SPEEDS = {1: 'slow', 2: 'normal', 3: 'fast'}
INTENSITIES = {0: 'fake', 1: 'slight', 2: 'medium', 3: 'hard'}
N_TRIALS_PER_CONDITION = 6
RATIO_FAKE_TO_ACTUAL = 1/3  # The number of fake perturbation trials relative to total actual perturbation trials
```

```python
subject_id = 'S05'
measurement_data = '20260210'
subject_identifier = '_'.join([subject_id, measurement_data])
speed = []
condition_permutated = []

# The speed values array for the whole measurement
# Each speed forms a block
n_trials_per_block = N_TRIALS_PER_CONDITION * (len(INTENSITIES) - 1) * (RATIO_FAKE_TO_ACTUAL + 1)
speed_values = np.repeat(list(SPEEDS.values()), n_trials_per_block)

# The number of repetitions for each condition within a block, in order of the intensity indices
n_repeat_condition = [N_TRIALS_PER_CONDITION * (len(INTENSITIES) - 1) * RATIO_FAKE_TO_ACTUAL] + [N_TRIALS_PER_CONDITION] * (len(INTENSITIES) - 1)

# Generate randomized condition order within each block
for i in range(len(SPEEDS)):
    condition_values = np.repeat(list(INTENSITIES.values()), n_repeat_condition)
    condition_permutated = np.append(condition_permutated, np.random.permutation(condition_values))

df = pd.DataFrame({
    'Overall Trial No.': np.arange(1, len(speed_values) + 1),
    'Block Speed': speed_values,
    'Block Trial No.': np.tile(np.arange(1, sum(n_repeat_condition) + 1), len(SPEEDS)),
    'Intensity': condition_permutated,
    'Invalid Trials': [None] * len(speed_values),
    'Note': [''] * len(speed_values)
})

subject_dir = os.path.join(root_dir, subject_identifier)
path = os.path.join(subject_dir, subject_identifier + '_trial_list.csv')
if not os.path.exists(subject_dir):
    os.makedirs(subject_dir)
if not os.path.exists(path):
    # pd.io.formats.excel.ExcelFormatter.header_style = None
    # with pd.ExcelWriter(path) as writer:
    #     for i in range(len(SPEEDS)):
    #         df.iloc[i*n_trials_per_block:(i+1)*n_trials_per_block,:].to_excel(writer, sheet_name=cur_subject+'_Block'+str(i+1), index=False)
    df.to_csv(path, index=False)
else:
    raise FileExistsError(f"File {path} already exists. Please check before overwriting.")
```
